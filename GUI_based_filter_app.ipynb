```python
import tkinter
import cv2
import PIL.Image, PIL.ImageTk
from tkinter import filedialog
import numpy as np
from imageio import imread
from skimage.filters import difference_of_gaussians
from tkinter import *


class App:
    def __init__(self, window, window_title):
        self.window = window                                    #Initializing
        self.window.title(window_title)
        
        self.image_path = filedialog.askopenfilename()          #load image
  
         # Load an image using OpenCV
        #cv_img = cv2.cvtColor(cv2.imread(self.image_path), cv2.COLOR_BGR2RGB)    #store the image in self.cv_img
        
                                                       #Resize function used to resize image to 512X512 dimension
        img = cv2.imread(self.image_path)
        img_sh = imread(self.image_path)
        dim = (512,512)
        if np.ndim(img_sh) == 3:
            resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
            self.cv_img = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
        elif np.ndim(img_sh) == 2:
            self.cv_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
                
                 # Get the image dimensions (OpenCV stores image data as NumPy ndarray)
        self.height, self.width, no_channels = self.cv_img.shape
                
        
        #Creating labels
        
        label1 = tkinter.Label(window,text="Filters")
        label1.grid(row=0,column=10, columnspan=1)
        
        label2 = tkinter.Label(window,text="Statistical Type")
        label2.grid(row=3,column=10, columnspan=1)
        
        label3 = tkinter.Label(window,text="Derivative Type")
        label3.grid(row=5,column=10, columnspan=1)
        
        label4 = tkinter.Label(window,text="Linear Type")
        label4.grid(row=3,column=18, columnspan=1)
        
        label5 = tkinter.Label(window,text="Histogram Equalization")
        label5.grid(row=0,column=18, columnspan=1)
 
         # Create a canvas that can fit the above image
        self.canvas = tkinter.Canvas(window, width = self.width, height = self.height)
        self.canvas.grid(row=0, column=1, rowspan=15, columnspan=5)

       # Use PIL (Pillow) to convert the NumPy ndarray to a PhotoImage
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.cv_img))

         # Add a PhotoImage to the Canvas
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
        #Creating scales for different variables
        
        self.w2 = Scale(window, from_=0, to=255,orient = HORIZONTAL, label='Threshold')
        self.w2.grid(row=4,column=20)
        
        self.past = 1   
        def fix(n):                                         #function for odd value of scale
            n = int(n)
            if not n % 2:
                self.kernel.set(n+1 if n > self.past else n-1)
                self.past = self.kernel.get() 
                
        self.kernel = Scale(from_=1, to_=15, command=fix, orient=HORIZONTAL, label='Kernel Size')
        self.kernel.grid(row=4, column=15)
        
        def fix_l(n):
            n = int(n)
            if not n % 2:
                self.kernel_lap.set(n+1 if n > self.past else n-1)
                self.past = self.kernel_lap.get()
                
        self.kernel_lap = Scale(from_=1, to_=15, command=fix_l, orient=HORIZONTAL, label='Kernel Size')
        self.kernel_lap.grid(row=6, column=15)
        
        def fix_sob(n):
            n = int(n)
            if not n % 2:
                self.kernel_sob.set(n+1 if n > self.past else n-1)
                self.past = self.kernel_sob.get()
                
        self.kernel_sob = Scale(from_=1, to_=23, command=fix_sob, orient=HORIZONTAL, label='Kernel Size')
        self.kernel_sob.grid(row=9, column=15)
        
        self.dir_X = Scale(from_=0, to_=1, orient=HORIZONTAL, label='dx')
        self.dir_X.grid(row=9, column=16)
        
        self.dir_Y = Scale(from_=0, to_=1, orient=HORIZONTAL, label='dy')
        self.dir_Y.grid(row=9, column=17)
        
        def fix_lg(n):
            n = int(n)
            if not n % 2:
                self.kernel_lg.set(n+1 if n > self.past else n-1)
                self.past = self.kernel_lg.get()
                
        self.kernel_lg = Scale(from_=1, to_=23, command=fix_lg, orient=HORIZONTAL, label='Kernel Size')
        self.kernel_lg.grid(row=7, column=15)
        
        self.kernel_X = Scale(from_=0, to_=1, orient=HORIZONTAL, label='SigmaX')
        self.kernel_X.grid(row=7, column=16)
        
        self.kernel_Y = Scale(from_=0, to_=1, orient=HORIZONTAL, label='SigmaY')
        self.kernel_Y.grid(row=7, column=17)
        
        def fix_low(n):
            n = int(n)
            if not n % 2:
                self.kernel_low.set(n+1 if n > self.past else n-1)
                self.past = self.kernel_low.get()
                
        self.kernel_low = Scale(from_=1, to_=23, command=fix_low, orient=HORIZONTAL, label='Kernel Size')
        self.kernel_low.grid(row=5, column=20)
        
        def fix_high(n):
            n = int(n)
            if not n % 2:
                self.kernel_high.set(n+1 if n > self.past else n-1)
                self.past = self.kernel_high.get()
                
        self.kernel_high = Scale(from_=1, to_=23, command=fix_high, orient=HORIZONTAL, label='Kernel Size')
        self.kernel_high.grid(row=6, column=20)
        
        self.sigma_h = Scale(from_=1, to_=10, orient=HORIZONTAL, label='Sigma')
        self.sigma_h.grid(row=6, column=21)
        
        self.sigma_hp = Scale(from_=10, to_=30, orient=HORIZONTAL, label='Sigma_high')
        self.sigma_hp.grid(row=7, column=21)
        
        self.sigma_lp = Scale(from_=1, to_=10, orient=HORIZONTAL, label='Sigma_low')
        self.sigma_lp.grid(row=7, column=20)
        
        self.dir_a = Scale(from_=0, to_=1, orient=HORIZONTAL, label='dir_X')
        self.dir_a.grid(row=8, column=15)
        
        self.dir_b = Scale(from_=0, to_=1, orient=HORIZONTAL, label='dir_Y')
        self.dir_b.grid(row=8, column=16)
        
        self.low_reject = Scale(from_=1, to_=7, orient=HORIZONTAL, label='lower cut-off')
        self.low_reject.grid(row=8, column=20)
        
        self.high_reject = Scale(from_=7, to_=20, orient=HORIZONTAL, label='higher cut-off')
        self.high_reject.grid(row=8, column=21)
        
        self.clip = Scale(from_=1, to_=15, orient=HORIZONTAL, label='Cliplimit')
        self.clip.grid(row=2, column=20)
        
        self.grid = Scale(from_=1, to_=15, orient=HORIZONTAL, label='Gridsize')
        self.grid.grid(row=2, column=21)
        
 
         # Button that lets the user apply filter to the image
    
        self.btn_load=tkinter.Button(window, text="color", width=15, command=self.load_image)
        self.btn_load.grid(row=1, column=13)
        
        self.btn_gray=tkinter.Button(window, text="gray", width=15, command=self.gray_image)
        self.btn_gray.grid(row=2, column=13)
        
        self.btn_median=tkinter.Button(window, text="median", width=15, command=self.median)
        self.btn_median.grid(row=4, column=13)
        
        self.btn_laplacian =tkinter.Button(window, text="laplacian", width=15, command=self.laplacian)
        self.btn_laplacian.grid(row=6, column=13)
        
        self.btn_lap_gaauss=tkinter.Button(window, text="laplace_gauss", width=15, command=self.lap_gauss)
        self.btn_lap_gaauss.grid(row=7, column=13)
        
        self.btn_prewitt=tkinter.Button(window, text="Prewitt", width=15, command=self.prewitt)
        self.btn_prewitt.grid(row=8, column=13)
        
        self.btn_sobel=tkinter.Button(window, text="sobel", width=15, command=self.sobel)
        self.btn_sobel.grid(row=9, column=13)
        
        
        self.btn_thresh=tkinter.Button(window, text="thresh", width=15, command=self.binary_thresh)
        self.btn_thresh.grid(row=4, column=19)
        
        self.btn_lpf=tkinter.Button(window, text="Low Pass", width=15, command=self.LPF)
        self.btn_lpf.grid(row=5, column=19)
        
        self.btn_hp=tkinter.Button(window, text="High Pass", width=15, command=self.high_pass)
        self.btn_hp.grid(row=6, column=19)
        
        self.btn_bp=tkinter.Button(window, text="Bandpass", width=15, command=self.bandpass)
        self.btn_bp.grid(row=7, column=19)
        
        
        self.btn_br=tkinter.Button(window, text="Bandreject", width=15, command=self.bandreject)
        self.btn_br.grid(row=8, column=19)
        
        
        self.btn_hist=tkinter.Button(window, text="global_hist", width=15, command=self.global_hist)
        self.btn_hist.grid(row=1, column=19)
        
        self.btn_cla=tkinter.Button(window, text="Adaptive", width=15, command=self.adaptive)
        self.btn_cla.grid(row=2, column=19)
 
        self.window.mainloop()

    #Function for all the filters
    
    def resize(self, image_path):
        img = cv2.imread(image_path)
        img_sh = imread(image_path)
        dim = (512,512)
        if np.ndim(img_sh) == 3:
            resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
            self.cv_img = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
        elif np.ndim(img_sh) == 2:
            self.cv_img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    
    def LPF(self):                                         #Low pass filter                           
        p = round(self.kernel_low.get())                   #getting values from scale
        self.cv_img = cv2.blur(self.cv_img, (p, p))
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.cv_img))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def load_image(self):                                 #color image
        self.resize(self.image_path)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.cv_img))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def gray_image(self):                                #gray image
        self.resize(self.image_path)
        img_sh = imread(self.image_path)
        img = self.cv_img
        if np.ndim(img_sh) == 3:                        #If channel is 3 then convert it to gray, if channel is 1, no change
            self.cv_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.cv_img))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
            
        elif np.ndim(img_sh) == 2:
            self.resize(self.image_path)
            img = self.cv_img
            img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(img_1))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)

    def binary_thresh(self):                           #binary threshold filter
        #Adding scale
        p = round(self.w2.get())
        self.resize(self.image_path)
        self.cv_img_t = cv2.cvtColor(self.cv_img, cv2.COLOR_BGR2GRAY)
        ret, self.cv_img = cv2.threshold(self.cv_img_t,p,255,cv2.THRESH_BINARY)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.cv_img))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)  
        
    def median(self):                                 #median
        p = round(self.kernel.get())
        self.cv_img = cv2.medianBlur(self.cv_img,p)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.cv_img))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def laplacian(self):                            #laplacian
        p = round(self.kernel_lap.get())
        self.resize(self.image_path)
        img = self.cv_img
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        laplacian = cv2.Laplacian(img_gray,cv2.CV_64F, ksize = p)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(laplacian))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def sobel(self):                                 #sobel
        p = round(self.kernel_sob.get())
        q = round(self.dir_X.get())
        r = round(self.dir_Y.get())
        if (q==0) & (r==0):
            print('Please select an option from dx and dy')
            self.resize(self.image_path)
            img = self.cv_img
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(img))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        else:
            self.resize(self.image_path)
            img = self.cv_img
            img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            sobel = cv2.Sobel(img_gray,-1,  dx=q, dy=r, ksize=p, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(sobel))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def lap_gauss(self):
        p = round(self.kernel_lg.get())                   #Laplacian of Gaussian
        q = round(self.kernel_X.get())
        r = round(self.kernel_Y.get())
        image2= cv2.GaussianBlur(self.cv_img,(p,p), sigmaX = q, sigmaY = r, borderType = cv2.BORDER_DEFAULT)
        src_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
        image1=cv2.Laplacian(src_gray,cv2.CV_64F,ksize=p)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(image1))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def prewitt(self):                                   #prewitt
        p = round(self.dir_a.get())
        q = round(self.dir_b.get())
        kernelx = p*np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
        kernely = q*np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
        self.resize(self.image_path)
        img = self.cv_img
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img_gaussian = cv2.GaussianBlur(gray,(3,3),0)
        img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
        img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)
        img_prewit=img_prewittx+img_prewitty
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(img_prewit))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
    
    def global_hist(self):                           #Global histogram, if 3 channels then split and apply equalization individually and then merge the channels
        self.resize(self.image_path)
        img_sh = imread(self.image_path)
        img = self.cv_img
        if np.ndim(img_sh) == 3:
            blue, green, red = cv2.split(img)
            img1_b =cv2.equalizeHist(blue)
            img1_g =cv2.equalizeHist(green)
            img1_r =cv2.equalizeHist(red)
            rgb_equ = cv2.merge([img1_r,img1_g,img1_b])
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(rgb_equ))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
            
        elif np.ndim(img_sh) == 2:
            self.resize(self.image_path)
            img = self.cv_img
            img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            equ = cv2.equalizeHist(img_1)
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(equ))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def adaptive(self):                                 #CLAHE
        self.resize(self.image_path)
        img_sh = imread(self.image_path)
        img = self.cv_img
        p = round(self.clip.get())
        q = round(self.grid.get())
        if np.ndim(img_sh) == 3:
            blue, green, red = cv2.split(img)
            clahe = cv2.createCLAHE(clipLimit = p, tileGridSize=(q,q))
            clahe_img1_b =clahe.apply(blue)
            clahe_img1_g =clahe.apply(green)
            clahe_img1_r =clahe.apply(red)
            rgb_clahe = cv2.merge([clahe_img1_r,clahe_img1_g,clahe_img1_b])
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(rgb_clahe))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
            
        elif np.ndim(img_sh) == 2:
            self.resize(self.image_path)
            img = self.cv_img
            img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            clahe = cv2.createCLAHE(clipLimit = 3.0, tileGridSize=(8,8))
            clahe_img = clahe.apply(img_1)
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(clahe_img))
            self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def high_pass(self):                             #High pass
        p = round(self.kernel_high.get())
        q = round(self.sigma_h.get())
        sigma = q
        image_hp = self.cv_img - cv2.GaussianBlur(self.cv_img, (p,p), sigma) + 127
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(image_hp))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def bandpass(self):                            #Band Pass
        p = round(self.sigma_hp.get())
        q = round(self.sigma_lp.get())
        self.resize(self.image_path)
        img = self.cv_img
        cv_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        filtered_image = difference_of_gaussians(cv_img, q, p)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(filtered_image*255))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)
        
    def bandreject(self):                        #Band Reject
        p=round(self.low_reject.get())
        q=round(self.high_reject.get())
        self.resize(self.image_path)
        img = self.cv_img
        img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        filt_image= img_1-255*difference_of_gaussians(img_1, p, q)
        self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(filt_image))
        self.canvas.create_image(0, 0, image=self.photo, anchor=tkinter.NW)  
        
        
 # Create a window and pass it to the Application object
App(tkinter.Tk(), "Tkinter and OpenCV")
```
